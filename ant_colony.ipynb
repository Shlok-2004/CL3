```python
# Ant Colony Optimization for TSP

import numpy as np
import random

# Distance Matrix
dist = np.array([
    [0, 6, 5, 7],
    [2, 0, 4, 8],
    [5, 4, 7, 3],
    [7, 8, 3, 0]
])

n = len(dist)
pheromone = np.ones((n, n))

best_path = []
best_cost = 999

for _ in range(10):

    path = [0]
    visited = {0}

    while len(path) < n:

        current = path[-1]

        choices = [i for i in range(n) if i not in visited]

        next_city = max(choices,
                        key=lambda x:
                        pheromone[current][x] / dist[current][x])

        path.append(next_city)
        visited.add(next_city)

    # Calculate Cost
    cost = sum(dist[path[i]][path[i+1]]
               for i in range(n-1))

    cost += dist[path[-1]][path[0]]

    # Update Best
    if cost < best_cost:
        best_cost = cost
        best_path = path

    # Update Pheromone
    for i in range(n-1):
        pheromone[path[i]][path[i+1]] += 1 / cost

print("Best Path:", best_path)
print("Best Cost:", best_cost)
```
