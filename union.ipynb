# Implement Union, Intersection, Complement and Difference operations on fuzzy sets. Also create fuzzy relations by Cartesian product of any two fuzzy sets and perform max-min composition on any two fuzzy relations.

```python
import numpy as np

# Fuzzy Sets
A = np.array([0.2, 0.5, 0.7])
B = np.array([0.4, 0.3, 0.8])

# Operations
print("Union:", np.maximum(A, B))
print("Intersection:", np.minimum(A, B))
print("Complement of A:", 1 - A)
print("Difference A-B:", np.minimum(A, 1 - B))

# Cartesian Product Relations
R = np.minimum.outer(A, B)
C = np.array([0.1, 0.9, 0.5])
S = np.minimum.outer(B, C)

print("\nRelation R:\n", R)
print("\nRelation S:\n", S)

# Max-Min Composition
T = np.array([[max(min(R[i][k], S[k][j])
          for k in range(len(B)))
          for j in range(len(C))]
          for i in range(len(A))])

print("\nMax-Min Composition:\n", T)
```
