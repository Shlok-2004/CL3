```python
import random

def fitness(x):
    return x**2

population = [random.uniform(-5, 5) for _ in range(10)]

generations = 20

best_solution = None
best_fitness = float('inf')

for gen in range(generations):

    population.sort(key=fitness)

    best = population[0]

    # Store global best
    if fitness(best) < best_fitness:
        best_solution = best
        best_fitness = fitness(best)

    print(f"Generation {gen+1}: Best Solution = {best:.5f}, Fitness = {fitness(best):.5f}")

    clones = [best for _ in range(5)]

    mutated = []
    for c in clones:
        mutated.append(c + random.uniform(-0.5, 0.5))

    population = mutated + population[:5]

print("\nFinal Best Solution:", best_solution)
```
