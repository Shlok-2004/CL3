```python
# Simple Artificial Immune System for Structure Damage Classification

import numpy as np
import matplotlib.pyplot as plt

# Generate Random Data
X = np.random.rand(20, 2)

# Labels: 1 = Damaged, 0 = Safe
y = ((X[:,0] > 0.5) & (X[:,1] > 0.5)).astype(int)

# Safe Data
safe = X[y == 0]

detectors = []

# Create Detectors
for i in range(5):

    d = np.random.rand(2)

    # Detector should stay away from safe data
    if all(np.linalg.norm(d - s) > 0.3 for s in safe):
        detectors.append(d)

# Test Structure
test = np.array([0.85, 0.75])

# Prediction
damage = False

for d in detectors:

    if np.linalg.norm(test - d) < 0.3:
        damage = True

# Output
print("Test Structure:", test)

if damage:
    print("Structure is Damaged")
else:
    print("Structure is Safe")

# -------- Visualization --------

plt.scatter(safe[:,0], safe[:,1], c='blue', label='Safe')
plt.scatter(X[y==1][:,0], X[y==1][:,1], c='red', label='Damaged')

detectors = np.array(detectors)

plt.scatter(detectors[:,0], detectors[:,1],
            c='green', marker='x', label='Detectors')

plt.scatter(test[0], test[1],
            c='black', marker='*', s=150, label='Test')

plt.legend()
plt.title("Structure Damage Classification")
plt.show()
```
