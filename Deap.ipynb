```python
# Simple Distributed Evolutionary Algorithm using DEAP

from deap import base, creator, tools
import random

# Minimize fitness
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.FitnessMin)

toolbox = base.Toolbox()

# Random float genes
toolbox.register("gene", random.uniform, -1, 1)

# Individual with 3 genes
toolbox.register("individual", tools.initRepeat,
                 creator.Individual, toolbox.gene, 3)

toolbox.register("population", tools.initRepeat,
                 list, toolbox.individual)

# Fitness function
def evaluate(ind):
    return (sum(x**2 for x in ind),)

toolbox.register("evaluate", evaluate)

# Create populations
pop1 = toolbox.population(n=5)
pop2 = toolbox.population(n=5)

# Evolution
for gen in range(10):

    for pop in [pop1, pop2]:

        for ind in pop:
            ind.fitness.values = toolbox.evaluate(ind)

        pop[:] = tools.selBest(pop, 5)

        # Mutation
        for ind in pop:
            ind[0] += random.uniform(-0.1, 0.1)

    # Migration
    pop1[0], pop2[0] = pop2[0], pop1[0]

# Best Solution
best = tools.selBest(pop1 + pop2, 1)[0]

print("Best individual:", best)
print("Best fitness:", best.fitness.values[0])
```
